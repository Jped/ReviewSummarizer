```python
nltk.pos_tag(['I','&', 'quality', 'food'])
```

```python
pos_list = {"JJ":5, "JJR":5, "JJS":5, "RB":5, "VBG":5, 
            "NN":0.5, "NNS":0.5, "NNP":0.5, "NNPS":0.5, 
            "PRP":0.5, "PRP$":0.5, "CC":0.5}
```

```python
stop = set(stopwords.words('english')) 
stop.add("&")
```

```python
tf_dict_array = []

for i, review in enumerate(reviews):
    review_dict = {}
    
    for word in review.split():
        if word in review_dict and word not in stop:
            review_dict[word] += 1
        elif word not in stopwords:
            review_dict[word] = 1
    tf_dict_array.append(review_dict)

idf_dict = {}

for tf_dict in  tf_dict_array:
    for word in tf_dict.keys():
        if word in idf_dict:
            idf_dict[word] += 1
        else:
            idf_dict[word] = 1
            
for word in idf_dict.keys():
    pos = nltk.pos_tag([word])[0][1]
    idf_dict[word] = math.log(len(reviews)/idf_dict[word])
    if pos in pos_list: 
        idf_dict[word] *= pos_list[pos]
    
tf_idf_arr = []

for tf_dict in tf_dict_array:
    cur_dict = {}
    
    for word in tf_dict.keys():
        cur_dict[word] = tf_dict[word]*idf_dict[word]
        
    tf_idf_arr.append(cur_dict)
    
tf_idf_arr
```

```python
bigrams_tfidf_arr = []

for i, review in enumerate(reviews):
    cur_dict = {}
    bigrams = nltk.bigrams(review.split())
    
    for bigram in bigrams:
        if bigram[0] not in stop and bigram[1] not in stop: 
            bikey = "{} {}".format(bigram[0],bigram[1])
            cur_dict[bikey] = tf_idf_arr[i][bigram[0]] + tf_idf_arr[i][bigram[1]]
    bigrams_tfidf_arr.append(cur_dict)
```

```python
sorted(bigrams_tfidf_arr[3],key = lambda x: bigrams_tfidf_arr[3][x])
```

```python
trigrams_tfidf_arr = []

for i, review in enumerate(reviews):
    cur_dict = {}
    trigrams = nltk.trigrams(review.split())
    
    for trigram in trigrams:
        if trigram[0] not in stop and trigram[1] not in stop and trigram[2] not in stop: 
            trikey = "{} {} {}".format(trigram[0],trigram[1],trigram[2])
            cur_dict[trikey] = tf_idf_arr[i][trigram[0]] + tf_idf_arr[i][trigram[1]] + tf_idf_arr[i][trigram[2]]
    trigrams_tfidf_arr.append(cur_dict)
```
